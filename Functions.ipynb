5. What are iterators in Python and how do they differ from iterables?
   -  iterators and iterables are both related to the concept of traversing data, but they have different roles and behaviors.
   - An iterable is any Python object that can return an iterator. An iterable is an object that can be looped over, and it contains a method __iter__() that returns an iterator.
   

```python
#Example for iterable
my_list = [1, 2, 3, 4]
# my_list is an iterable because it can be passed to iter()
iterator = iter(my_list)  # Returns an iterator for the list

# Iterating over an iterable
for item in my_list:
    print(item)
```

An iterator is an object that represents a stream of data. It allows you to traverse through all the elements of a collection (like a list, tuple, etc.) one at a time. Iterators are created from iterables, and they maintain the state of the current position in the iteration.

```python
#Example for iterator
my_list = [1, 2, 3, 4]
iterator = iter(my_list)  # Create an iterator from the list

print(next(iterator))
print(next(iterator))
print(next(iterator))
print(next(iterator))
# print(next(iterator))
```

6. Explain the concept of generators in Python and how they are defined?
   - A generator is a special type of iterator that allows you to create iterators in a simple and memory-efficient way. Unlike regular functions that return a single value using the return statement, generators use the yield keyword to return a value, but they maintain their state between function calls, allowing them to generate a sequence of values lazily, one at a time.
   Defining a Generator:A generator is defined just like a normal function, except it uses the yield keyword instead of return to send values back to the caller.

```python
#EX:
def count_up_to(max):
    count = 1
    while count <= max:
        yield count  # Yield the current value of count
        count += 1

# Creating a generator instance
counter = count_up_to(3)

# Using next() to get values from the generator
print(next(counter))
print(next(counter))
print(next(counter))
# print(next(counter))
```

9. Explain the purpose and usage of the `map()` function in Python?
   - The map() function in Python is a built-in higher-order function used to apply a given function to all the items in an iterable (such as a list, tuple, or any other iterable) and return an iterator that produces the results. It allows you to transform or modify data in a sequence without needing to write an explicit loop.
   

```python
#EX:we will use map() to square each element in a list:
# Function to square a number
def square(x):
    return x * x

# List of numbers
numbers = [1, 2, 3, 4]

# Apply the square function to each item in the list
squared_numbers = map(square, numbers)

# Convert the map object to a list and print
print(list(squared_numbers))  # Output: [1, 4, 9, 16]
```

10.  What is the difference between `map()`, `reduce()`, and `filter()` functions in Python?
     - In Python, the functions map(), reduce(), and filter() are all built-in higher-order functions used for processing and transforming data in iterables (like lists, tuples, etc.). While they share some similarities in that they all operate on iterables, they each serve different purposes and are used in different scenarios.
     - map() Function:
      Purpose: The map() function applies a given function to all the items in an iterable (or multiple iterables) and returns an iterator that produces the results.
      output: An iterator that yields the results of applying the function to each item in the iterable.
      typical Use Case: Transforming or modifying data in a sequence (e.g., applying a mathematical operation to each element in a list).
                       

```python
#EX:
numbers = [1, 2, 3, 4]
squared = map(lambda x: x * x, numbers)
print(list(squared))
```

reduce() Function:
Purpose: The reduce() function applies a given function to the items of an iterable (or multiple iterables) cumulatively. It combines the items of the iterable into a single value by applying the function to the elements sequentially.
Output: A single value that is the result of applying the function cumulatively across the iterable.
Typical Use Case: Accumulating or combining data into a single value (e.g., calculating the sum, product, or any other accumulation operation).

```python
#EX:
from functools import reduce

numbers = [1, 2, 3, 4]
product = reduce(lambda x, y: x * y, numbers)
print(product)
```

filter() Function:
Purpose: The filter() function filters elements from an iterable by applying a given function that returns True or False (a boolean value) for each element. It retains only the elements where the function returns True.
Output: An iterator that yields the elements of the iterable for which the function returns True.
Typical Use Case: Filtering or selecting items from a sequence based on a condition.

```python
#EX:
numbers = [1, 2, 3, 4, 5]
even_numbers = filter(lambda x: x % 2 == 0, numbers)
print(list(even_numbers))
```

1.  Write a Python function that takes a list of numbers as input and returns the sum of all even numbers in
the list?

```python
def sum_of_even_numbers(numbers):
    # Use filter to select only even numbers and then sum them
    even_numbers = filter(lambda x: x % 2 == 0, numbers)
    return sum(even_numbers)

# Example usage:
numbers = [47, 11, 42, 13, 8, 30]
result = sum_of_even_numbers(numbers)
print("Sum of even numbers:", result)
```

```python
#4. Write a Python function that checks if a given number is prime or not from 1 to 200?
def is_prime(n):
    if n <= 1:
        return False  # Numbers less than or equal to 1 are not prime
    for i in range(2, int(n ** 0.5) + 1):
        if n % i == 0:
            return False  # If divisible by any number other than 1 and itself
    return True  # If no divisors found, the number is prime

# Check prime numbers from 1 to 200
prime_numbers = [num for num in range(1, 201) if is_prime(num)]

print("Prime numbers from 1 to 200:", prime_numbers)
```

```python
#5. Create an iterator class in Python that generates the Fibonacci sequence up to a specified number of
terms?
class FibonacciIterator:
    def __init__(self, num_terms):
        self.num_terms = num_terms  # Number of terms in the Fibonacci sequence
        self.current = 0  # First term
        self.next = 1  # Second term
        self.count = 0  # Counter to track the number of terms generated

    def __iter__(self):
        return self  # Returns the iterator object itself

    def __next__(self):
        if self.count >= self.num_terms:
            raise StopIteration  # Stop iteration when the desired number of terms is reached
        else:
            fib_number = self.current
            self.current, self.next = self.next, self.current + self.next  # Update the Fibonacci sequence
            self.count += 1
            return fib_number

# Example usage:
num_terms = 10  # Number of Fibonacci terms to generate
fib_iterator = FibonacciIterator(num_terms)

# Print Fibonacci sequence
for term in fib_iterator:
    print(term)
```

```python
#6. Write a generator function in Python that yields the powers of 2 up to a given exponent?
def powers_of_2(exponent):
    for i in range(exponent + 1):
        yield 2 ** i

# Example usage:
exponent = 5  # The highest exponent
for power in powers_of_2(exponent):
    print(power)
```

7. Implement a generator function that reads a file line by line and yields each line as a string?

```python
def read_file_line_by_line(file_path):
    with open(file_path, 'r') as file:
        for line in file:
            yield line.strip()  # .strip() removes leading/trailing whitespace
```

12. Write a Python program, which returns a list with 2-tuples. Each tuple consists of the order number and the
product of the price per item and the quantity. The product should be increased by 10,- € if the value of the
order is smaller than 100,00 €?

```python
def calculate_order_values(orders):
    order_values = []

    for order_number, quantity, price_per_item in orders:
        total_value = quantity * price_per_item  # Calculate the total value for the order
        if total_value < 100:
            total_value += 10  # Add 10 € if the order value is smaller than 100 €

        order_values.append((order_number, total_value))  # Add tuple (order_number, total_value)

    return order_values

# Example orders: Each order is represented as (order_number, quantity, price_per_item)
orders = [
    (1, 2, 25.50),  # Order 1: 2 items, 25.50 € each
    (2, 1, 110.00), # Order 2: 1 item, 110.00 € each
    (3, 3, 20.00),  # Order 3: 3 items, 20.00 € each
]

# Calculate the order values
order_values = calculate_order_values(orders)

# Output the results
print("Order values:")
for order in order_values:
    print(f"Order {order[0]}: {order[1]:.2f} €")
```
